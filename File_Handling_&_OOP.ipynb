# File Handling

```python
#Create and write a file
file = open("sample.txt", "w") #open file in write mode
file.write("Hello World\n")
file.write("Soumit Dey\n")
file.write("Student of UAP\n")
file.close()
```

```python
#Read a file
file=open("sample.txt","r") #open file in read mode
content=file.read()
print(content)
file.close()
```

```python
#add a new content in file
file=open("sample.txt","a")
file.write("Just write a new line\n")
file.close()
```

## File Handling using Context Manager

```python
with open("sample2.txt","w") as file:
    file.write("user1 : Soumit\n")
    file.write("user1 : Milton\n")
```

```python
with open("sample2.txt","r") as file:
    for line in file.readlines():
        print(line)
```

# Introduction to OOP

```python
# Class and Object
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_average(self):
        avg = sum(self.marks) / len(self.marks)
        return avg

    def print_result(self):
        avg1 = self.calculate_average()
        print(f"{self.name} Average: {avg1}")

```

```python
students = [
    Student("Soumit", [80, 75, 90]),
    Student("Milton", [60, 65, 70]),
    Student("Shakil", [80, 75, 90]),
    Student("Alex", [60, 65, 70]),
    Student("Mehedi", [60, 65, 70])
]

for i in students:
    i.print_result()
```
