# Introduction to OOP

```python
# Class and Object
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_average(self):
        avg = sum(self.marks) / len(self.marks)
        return avg

    def print_result(self):
        avg1 = self.calculate_average()
        print(f"{self.name} Average: {avg1}")

```

```python
students = [
    Student("Soumit", [80, 75, 90]),
    Student("Milton", [60, 65, 70]),
    Student("Shakil", [80, 75, 90]),
    Student("Alex", [60, 65, 70]),
    Student("Mehedi", [60, 65, 70])
]

for i in students:
    i.print_result()
```

## Inheritance

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

# Child classes
class Dog(Animal):
    pass

class Cat(Animal):
    pass

class Cow(Animal):
    pass

# Using the classes
dog = Dog("Max")
cat = Cat("Milo")
cow = Cow("Bessie")

dog.eat()
cat.eat()
cow.eat()
dog.name
```

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating. And it eats {self.food}.")

# Child class
class Dog(Animal):
    def __init__(self,name,food):
        super().__init__(name)
        self.food = food
        print("Dog created")


# Using the class
dog = Dog("Max","chicken")

dog.eat()
dog.name
```

## Encapsulation

```python
class ATM:
    def __init__(self,balance):
        self.__balance = balance # double underscore makes a property/attribute private
        
    def show_balance(self):
        print(f"The balance is {self.__balance}")
    
    def deposit(self,amount):
        self.__balance = self.__balance + amount
        return self.__balance
    
    def withdraw(self,amount):
        self.__balance = self.__balance - amount
        return self.__balance
        
        
b=ATM(500)
b.__balance # cannont access balance directly. since it is a private data
```

```python
b.show_balance()
b.deposit(200)
b.show_balance()
b.withdraw(50)
b.show_balance()
```

## Polymorphism (Small Project)

```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def calculate_average(self):
        return sum(self.marks)/len(self.marks)
    
    def print_result(self):
        avg = self.calculate_average()
        print(f"Name:{self.name} | Average:{avg}")
        
        
class OnlineStudent(Student):
    def __init__(self,name,marks,online_attendance):
        super().__init__(name,marks)
        self.online_attendance = online_attendance
        
    def final_calculation(self):
        avg_on = self.calculate_average()
        return avg_on*0.9 + self.online_attendance*0.1

    
class OfflineStudent(Student):  
    def __init__(self,name,marks,offline_attendance):
        super().__init__(name,marks)
        self.offline_attendance = offline_attendance
        
    def final_calculation(self):
        avg_off = self.calculate_average()
        return avg_off*0.7 + self.offline_attendance*0.3    
    
```

```python
students=[
    OnlineStudent("Max",[80,90],82),
    OnlineStudent("Lucas",[85,75],90),
    OfflineStudent("El",[70,60],63),
    OfflineStudent("Mike",[75,58],64),
    OfflineStudent("Dustin",[90,95],85)
]
for i in students:
    print(f"Name: {i.name} | Final Score: {i.final_calculation()}")
    #i.print_result()
```
