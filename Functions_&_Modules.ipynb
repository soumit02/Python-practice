```python
#return value--> 1. It can be passed to another function
#                2.Essential for real-world software logic
def add(a,b):
    return a+b
    
def Even_func(sum1):
    if sum1%2==0:
        print("Even sum.")

def Odd_func(sum2):
    if sum2%2!=0:
        print("Odd sum.")
         
a=int(input("Enter 1st number: "))
b=int(input("Enter 2nd number: "))
sum=add(a,b)
print(f"Sum is {sum}.")
Even_func(sum)
Odd_func(sum)
```

```python
def calculate_final_price(price, tax_rate):
    total = price + price*tax_rate
    print(f"Total: {total}")
    return total

# postion doesn't matter  -> keyword arguments
calculate_final_price(price=3000, tax_rate=0.15)
calculate_final_price(tax_rate=0.15, price=3000)
```

# Error Handling with Try & Except

```python
def divide(a,b):
    try:
        result=a/b
        return result
    except ZeroDivisionError:
        return "Error: You cannot divide by zero."
    except TypeError:
        return "Error: Try again with valid input."
divide(5,"a")    
```

```python
def divide(a,b):
    try:
        result=a/b
        return result
    except Exception as e:
        print(f"Error: {e}")
divide(0,0) 
```

```python
def divide(a,b):
    try:
        return a/b
    except Exception as e:
        return e
    
try:    
    print(divide())
except Exception as e:
    print(f"Error: {e}")
```
